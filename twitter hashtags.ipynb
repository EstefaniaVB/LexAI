# Count hashtags per topic

```python
q='data'
params=dict(q='data',limit=100000)
headers={'X-Meili-API-Key':'OTkwNzQ0ZGRkZTc0NDcwM2RlMzFlOGIx'}

lexai_url = f"http://35.223.18.2/indexes/twitter_query/search/"
data = requests.get(lexai_url,params=params,headers=headers).json()
data_query_df=pd.DataFrame(data['hits'])

lexai_url = f"http://35.223.18.2/indexes/twitter_press/search/"
data = requests.get(lexai_url,params=params,headers=headers).json()
data_press_df=pd.DataFrame(data['hits'])

lexai_url = f"http://35.223.18.2/indexes/twitter_politicians/search/"
data = requests.get(lexai_url,params=params,headers=headers).json()
data_polit_df=pd.DataFrame(data['hits'])
```

```python
data_query_df.shape, data_press_df.shape, data_polit_df.shape
```

```python
hashtags=[]
for i in data_query_df['hashtags']:
    if i !='':
        for j in i.lower().split(', '):
            hashtags.append(j)
for i in data_press_df['hashtags']:
    if i !='':
        for j in i.lower().split(', '):
            hashtags.append(j)
for i in data_polit_df['hashtags']:
    if i !='':
        for j in i.lower().split(', '):
            hashtags.append(j)
```

```python
hashtags
```

```python
import collections
counter=collections.Counter(hashtags)
```

```python
dict(counter.most_common(10)).keys()
```
