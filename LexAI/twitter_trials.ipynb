```python
import os
import re
from datetime import datetime
from os.path import dirname, join
from time import mktime, sleep
from unicodedata import normalize

import pandas as pd
from dotenv import load_dotenv
from google.cloud import translate
from twython import Twython
```

```python
for result in tweets:
    dict_ = {}

    dict_['id'] = result['id']
    dict_['user'] = result['user']['screen_name']
    text = normalize('NFKD', result['full_text'])
    dict_['text'] = re.sub(r'https?:\/\/\S*', '', text, flags=re.MULTILINE)
    #dict_['text_en'] = self.gtrans(dict_['text'], dest='en')

    dt = datetime.strptime(result['created_at'],
                           "%a %b %d %H:%M:%S %z %Y")
    dict_['date'] = datetime.strftime(dt, "%Y/%m/%d %H:%M:%S")
    dict_['lang'] = result['lang']
    dict_['iso_lang'] = result.get('metadata', {}).get('iso_language_code', None)

    dict_['user_verified'] = result['user']['verified']
    dict_['followers_count'] = result['user']['followers_count']
    dict_['user_loc'] =  normalize('NFKD', result['user']['location'])
    dict_['user_desc'] = normalize('NFKD', result['user']['description'])
    #dict_['user_desc_en'] = self.gtrans(dict_['user_desc'], dest='en')
    dict_['user_image'] = result['user']['profile_image_url']

    dict_['hashtags'] = ', '.join([i['text'] 
                                   for i in result['entities']['hashtags']])
    dict_['mentions'] = ', '.join([i['screen_name'] 
                                   for i in result['entities']['user_mentions']])
    dict_['retweet_count'] = result['retweet_count']
    dict_['favorite_count'] = result['favorite_count']

    dict_['timestamp'] = mktime(dt.timetuple())
    dict_['link'] = f"https://twitter.com/{dict_['user']}/status/{dict_['id']}"

dict_
```
