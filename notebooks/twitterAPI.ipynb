### stats for tweets in a profile

```python
number_of_tweets = 50
tweets = []
likes = []
time = []

for i in tweepy.Cursor(api.user_timeline, id='euronews', tweet_mode="extended").items(number_of_tweets):
    tweets.append(i.full_text)
    likes.append(i.favorite_count)
    time.append(i.created_at)
```

```python
tweets
```

```python
df = pd.DataFrame({'tweets':tweets, 'likes':likes, 'time':time})
```

```python
df
```

sorting out retweets:

```python
df = df[~df.tweets.str.contains("RT")]
```

```python
df
```

```python
df = df.reset_index(drop=True)
```

get the most liked tweets:

```python
mostlike = df.loc[df.likes.nlargest(5).index]
```
